```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("summer.csv")
```

##1. In how many cities Summer Olympics is held so far?

```python
print("The number of countries where Summer Olympics were held are", len(df['City'].unique()))
```

#2. Which sport is having most number of Gold Medals so far? (Top 5)


```python
data=[]
df2 = df[ (df.Medal == 'Gold')]
for Sport in df2['Sport'].unique():
  data.append([Sport, len(df2[df2['Sport']==Sport])])

data=pd.DataFrame(data,columns=['Sport','Number of gold medals'])

data.sort_values(by='Number of gold medals',ascending=False).head().plot(x='Sport',y='Number of gold medals', kind='bar',figsize=(10,10))
```

#4. Which player has won most number of medals? (Top 5

```python
data3=[]
for Athlete in df['Athlete'].unique(): 
  data3.append([Athlete, len(df[df['Athlete']==Athlete])])

data3=pd.DataFrame(data3,columns=['Athlete','Number of medals'])

data3.sort_values(by='Number of medals',ascending=False).head().plot(x='Athlete',y='Number of medals', kind='bar',figsize=(10,10))



```

#5. Which player has won most number Gold Medals of medals? (Top 5)

#6. In which year India won first Gold Medal in Summer Olympics?

```python
india = df[(df.Country=="IND") & (df.Medal == 'Gold')]
print("India won the first Gold medal in", india.Year.head(1)) 
```

#7. Which event is most popular in terms on number of players? (Top 5)

```python
data5=[]
for event in df['Event'].unique():
    data5.append([event,len(df[df['Event']==event])])

data5=pd.DataFrame(data5, columns=['Event','Number of players']) 
data5.sort_values(by='Number of players',ascending=False).head().plot(x='Event',y='Number of players', kind='bar',figsize=(10,15))
```

#8. Which sport is having most female Gold Medalists? (Top 5)

```python
female = df[(df.Gender=="Women") & (df.Medal == 'Gold')]
data6=[]
for sport in female['Sport'].unique():
    data6.append([sport,len(female[female['Sport']==sport])])

data6=pd.DataFrame(data6, columns=['Sport','Number of female gold medalists']) 
data6.sort_values(by='Number of female gold medalists',ascending=False).head().plot(x='Sport',y='Number of female gold medalists', kind='bar',figsize=(10,15))
```
